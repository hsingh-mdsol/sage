```python
import pandas as pd
import requests
from tqdm import tqdm
import numpy as np
%load_ext autoreload
%autoreload 2
import nppes
```

# Import Data

```python
df = pd.read_csv('consolidated_kme_list_230317_for_vikas.csv')
df
```

# Get addresses

```python
nppes_add = nppes.NPPES()
npis = [x for x in df['PS/IQVIA_NPI'] if len(x) == 10]
address_df = pd.DataFrame()
for i in tqdm(npis):
    address_df = pd.concat([address_df, nppes_add.get_data(i)])
address_df.reset_index(inplace=True, drop=True)
```

```python
address_df = address_df.rename(columns={'npi':'PS/IQVIA_NPI'})
address_df
```

# Join onto full data

```python
df_final = df.merge(address_df, on='PS/IQVIA_NPI', how='left')
df_final
```

```python
# inserting - after first 5 digits of full zip codes
new_zips = []
for i in df_final['postal_code']:
    if i is not np.nan:
        if len(str(i)) > 5:
            new_zips.append(i[:5] + '-' + i[5:])
        else:
            new_zips.append(i)
    else:
        new_zips.append(i)
df_final['postal_code'] = new_zips
df_final
```

```python
df_final.to_csv('nppes_address_pull.csv', index=False)
```
